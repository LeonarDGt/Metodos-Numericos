```python
import math
xo=0
i=0
xmasuno=0
print ("{0:10}  {1:15}".format("Iteracion","   Valor Punto Fijo\n"));
while True:
    if i==0:
        xmasuno=0
    else:
        xmasuno=math.exp(-xo) 
        if xmasuno==xo:
            print ("{0:10} | {1:20}  {2:20}".format(i,xmasuno,"=> Esta es la raiz"));
            break
    xo=xmasuno
    print ("{0:10} | {1:20}".format(i,xmasuno));
    if i==100:
        break
    i=i+1
```
