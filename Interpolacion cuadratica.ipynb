```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline

b0=0;
b1=0;
b2=0;
bf=0;
xx=[];
x=[]
y=[]
z=[]
vx=[]
vy=[]
vz=[]
l1=0
l6=1.791759
l4=1.386294
l22=l1+((l4)/(4-1))*(2-1)
b1=(l4-l1)/(4-1)
b2=(((l6-l4)/(6-4))-b1)/(6-1)
l=1.0;
for l in range(1,36):
    x.append(math.log(l/4))
    xx.append(b0+b1*((l/4)-1)+b2*((l/4)-1)*((l/4)-4) );

z.append(l1) 
z.append(l22) 
z.append(l4) 
plt.title("interpolacion lineal")
plt.grid(True)
plt.axis([0,9,0,2.5])
plt.ylabel('x2')
plt.xlabel('x1')

for k in range(1,36):
    vx.append(k/4)
    vy.append(k/4)
    
for m in range(1,9):
    if (m==1):
        vz.append(m)
    if (m==2):
        vz.append(m)
    if (m==4):
        vz.append(m) 
plt.plot(vx,x,'g',linewidth=2,label='Valor verdadero')
plt.plot(vy,xx,'r--',linewidth=2,label='f(cuadratica)')
plt.plot(vz,z,'k--',linewidth=2,label='f(ln(4))')
plt.legend(loc = 4)
plt.show()
```
