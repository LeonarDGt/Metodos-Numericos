```python
import math
import matplotlib.pyplot as L
%matplotlib inline

g= 9.8
m= 68.1
c= 12.5
t=2
b=0
b1=0
Ev=0
Er=0
Ea=0
v3=0
e=[]
d=[]
x=[]
z=[]
y=[]
p=[]
print("         Tiempo  "+"     Analitico  "+"      Metodico"+"       Error Verdadero"+"     Error Relativo"+"    Error Aproximado\n");
for i in range(50):
    v=(g*m/c)*(1-pow(math.exp(1),-(c/m)*i*2))
    x.append(i)
    y.append(v)
#print("metodos numericos");
    if i==0:
       v1=0
    else:
        b1=b
        v1=b+(g-((c*b)/m))*(t)
        b=v1
    p.append(v1)
#print("Error verdadero");
    Ev=(v-v1)
#print("Error relativo");
    if i==0:
        Er=0
    else:
        Er=(Ev/v1)*100
#print("Error aproximado");
    if b==0:
        Ea=0
    else:
        Ea=((v1-b1)/v1)*100
    print ("{0:15} | {1:15} | {2:15} | {3:15}   | {4:15} | {5:15}".format((i*2),round(v,8),round(v1,8),round(Ev,8),round(Er,8),round(Ea,8)));
    z.append(Er) 
    d.append(Ev*(-1))
    
L.plot(x,y,'b',linewidth=2,label='Analitico')
L.hold(True)
L.grid(True)
L.plot(x,p,'k--',linewidth=2,label='Metodico')
L.plot(x,z,'b--',linewidth=2,label='E. Relativo')
L.plot(x,d,'g',linewidth=2,label='E. Verdadero')
L.legend(loc = 'lower right')
L.title("Caida Libre")
L.xlabel("Tiempo") 
L.ylabel("Velocidad") 
L.show()
```
